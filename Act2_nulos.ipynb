```python
#Carga desde un archivo .csv sin indice
df= pd.read_csv('Ventas_totales.csv')
df
```

```python
df.info()
```

```python
#Identificar valores nulos por columna
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
#Identificar valores nulos por dataframe
valores_nulos=df.isnull().sum().sum()
valores_nulos
```

```python
df1=df.copy()
```

```python
#salon_ventas cuenta con 6 datos nulos lo imputamos por la media para mantener la tendencia central del conjunto de datos.
df1["salon_ventas"]=df1["salon_ventas"].fillna(round(df1["salon_ventas"].mean(),1))
```

```python
#verificar que salon_ventas ya no tenga nulos
valores_nulos_salon_ventas = df1['salon_ventas'].isnull().sum()
valores_nulos_salon_ventas
```

```python
#Tarjetas de credito cuenta con 7 datos nulos lo imputamos por la media para mantener la tendencia central del conjunto de datos.
df1["tarjetas_credito"]=df1["tarjetas_credito"].fillna(round(df1["tarjetas_credito"].mean(),1))
```

```python
#verificar que tarjeta de credito ya no tenga nulos
valores_nulos_tarjeta_credito = df1['tarjetas_credito'].isnull().sum()
valores_nulos_tarjeta_credito
```

```python
#Tarjetas de debito cuenta con 1 dato nulo lo imputamos por la mediana dado que es el valor central del conjunto de datos.
df1["tarjetas_debito"]=df1["tarjetas_debito"].fillna(round(df1["tarjetas_debito"].median(),1))
```

```python
#verificar que tarjeta de debito ya no tenga nulos
valores_nulos_tarjeta_debito = df1['tarjetas_debito'].isnull().sum()
valores_nulos_tarjeta_debito
```

```python
#Otros medios cuenta con 1 valor nulo lo imputamos por la media para mantener la tendencia central del conjunto de datos.
df1["otros_medios"]=df1["otros_medios"].fillna(round(df1["otros_medios"].mean(),1))
```

```python
#verificar que otros_medios ya no tenga nulos
valores_nulos_tarjeta_debito = df1['otros_medios'].isnull().sum()
valores_nulos_tarjeta_debito
```

```python
#subtotal_ventas_alimentos_bebidas cuenta con 10 datos nulos lo imputamos por la media para mantener la tendencia central del conjunto de datos.
df1["subtotal_ventas_alimentos_bebidas"]=df1["subtotal_ventas_alimentos_bebidas"].fillna(round(df1["subtotal_ventas_alimentos_bebidas"].mean(),1))
```

```python
#verificar que subtotal_ventas_alimentos_bebidas ya no tenga nulos
valores_nulos_subtotal_ventas_alimentos_bebidas = df1['subtotal_ventas_alimentos_bebidas'].isnull().sum()
valores_nulos_subtotal_ventas_alimentos_bebidas
```

```python
#Bebidas cuenta con 1 dato nulo lo imputamos por ffill ya que rellena los valores faltantes
#utilizando el valor m√°s reciente disponible en la misma columna.
df1["bebidas"] =df1["bebidas"].fillna(method="ffill")
```

```python
#verificar que bebidas ya no tenga nulos
valores_nulos_bebidas = df1['bebidas'].isnull().sum()
valores_nulos_bebidas
```

```python
#Almacen cuenta con 2 datos nulo lo imputamos por bffill ya que rellena los valores faltantes
#utilizando el valor m√°s reciente pero despu√©s del valor faltante.
df1["almacen"] =df1["almacen"].fillna(method="bfill")
```

```python
#verificar que almacen ya no tenga nulos
valores_nulos_almacen = df1['almacen'].isnull().sum()
valores_nulos_almacen
```

```python
#verificar que lacteos ya no tenga nulos
valores_nulos_lacteos = df1['lacteos'].isnull().sum()
valores_nulos_lacteos
```
